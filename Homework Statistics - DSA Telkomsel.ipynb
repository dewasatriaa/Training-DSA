```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DSA\Statistics III\car_dataset.csv", encoding='latin1')
df.head(10)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df2 = df.drop(['Market Category'], axis=1)
```

```python
df3 = df2.dropna()
```

```python
df3.isnull().sum()
```

```python
def distplot(feature):
    plt.style.use('ggplot')
    plt.figure(figsize=(9, 5))
    plt.title("Displot for {}".format(feature))
    ax0 = sns.distplot(df[feature].dropna())
distplot('highway MPG')
distplot('city mpg')
distplot('MSRP')
distplot('Engine HP')
```

```python
df4 = df3.filter(['Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', 'Popularity', 'MSRP'],axis=1)
```

```python
from sklearn.preprocessing import RobustScaler
robust_scaler = RobustScaler()
df_robust = robust_scaler.fit_transform(df4)
```

```python
df4 = pd.DataFrame(df_robust)
```

```python
dict_df4 = {0: 'Engine HP', 1: 'Engine Cylinders', 2: 'highway MPG', 3: 'city mpg', 4: 'Popularity', 5: 'MSRP'}
df5 = df4.rename(dict_df4, axis=1)
```

```python
def distplot(feature):
    plt.figure(figsize=(9, 4))
    plt.title("Displot for {}".format(feature))
    ax0 = sns.distplot(df5[feature].dropna())
distplot('Engine HP')
distplot('Engine Cylinders')
distplot('highway MPG')
distplot('city mpg')
```

```python
g = sns.PairGrid(df5)
g = g.map(plt.scatter)
```
